```python
x_Axis=d1.drop('mpg',axis=1)
y_Axis=d1.mpg
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_Axis , y_Axis , test_size = 0.2 , random_state=30)
```

```python
#standard scaler transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#ML Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
#accuracy
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(pred,y_test)
r2=r2_score(pred,y_test)

print('mean_squared_error: ',mse)
print('r2_score: ',r2)
```

```python
df=pd.DataFrame({'Y_test':list(y_test),'Prediction':pred})
df
```

```python
sns.lmplot(x='Y_test',y='Prediction',data=df)
plt.show()
```
