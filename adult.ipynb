```python
#ip/op creation
ip=adult.drop('Listing of attributes',axis=1)
op=adult['Listing of attributes']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#standard scaler transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python

x_test
```

### linear model

```python
#ML Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
#accuracy
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(pred,y_test)
r2=r2_score(pred,y_test)

print('mean_squared_error: ',mse)
print('r2_score: ',r2)
```

```python
df=pd.DataFrame({'Y_test':list(y_test),'Prediction':pred})
df
```

```python
sns.lmplot(x='Y_test',y='Prediction',data=df)
plt.show()
```

```python
#ML Model
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_test
```

```python
pred=lr.predict(x_test)
pred
```

```python
#accuracy via confusion metrics
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
ac = accuracy_score(pred,y_test)
rc = recall_score(pred,y_test)
pc = precision_score(pred,y_test)
f1 = f1_score(pred,y_test)

print('Accuracy : ',ac)
print('Recall : ',rc)
print('Precision : ',pc)
print('F1 Score : ',f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(pred,y_test)
cm
```

```python
cmd=ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```
